```python
import seaborn as sns
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression,Lasso,Ridge
```

```python
import numpy as np
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split

# Load data
sales = pd.read_csv('/Users/siddiqkhan/Downloads/StudentsPerformance.csv')

# Features and target
X = sales.drop(columns=['writing score'])
y = sales['writing score']

# Categorical columns
categorical_cols = [
    'gender',
    'race/ethnicity',
    'parental level of education',
    'lunch',
    'test preparation course'
]

# ColumnTransformer
ct = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_cols)
    ],
    remainder='passthrough'
)

# Prepare X
X_prepared = ct.fit_transform(X)
print("Prepared shape:", X_prepared.shape)
print(type(X_prepared))
print(X_prepared)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_prepared, y, test_size = 0.2, random_state = 0)

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
print("Train R^2:", regressor.score(X_train, y_train))
print("Test R^2:", regressor.score(X_test, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor

# 1. Random Forest model
rf = RandomForestRegressor(
    n_estimators=200,
    random_state=0,
    n_jobs=-1
)

rf.fit(X_train, y_train)

print("=== Model Comparison ===")
print(f"Linear Regression - Train R^2: {regressor.score(X_train, y_train):.3f}")
print(f"Linear Regression - Test  R^2: {regressor.score(X_test, y_test):.3f}")
print()
print(f"Random Forest      - Train R^2: {rf.score(X_train, y_train):.3f}")
print(f"Random Forest      - Test  R^2: {rf.score(X_test, y_test):.3f}")
```

```python
def evaluate_model(true, predicted):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2_square = r2_score(true, predicted)
    return mae, rmse, r2_square
    
    
```

```python
models = {
    "Lasso": Lasso(),
    "Ridge": Ridge(),
    "KNN Regressor": KNeighborsRegressor(),
    "Decision Tree": DecisionTreeRegressor(random_state=0),
    "AdaBoost Regressor": AdaBoostRegressor(random_state=0)
}

results = []

for name, model in models.items():
    print("\n")
    print("Model:", name)
    print("------------------")

    # Train
    model.fit(X_train, y_train)

    # Predictions
    y_train_pred = model.predict(X_train)
    y_test_pred  = model.predict(X_test)

    # Evaluate
    model_train_mae, model_train_rmse, model_train_r2 = evaluate_model(y_train, y_train_pred)
    model_test_mae,  model_test_rmse,  model_test_r2  = evaluate_model(y_test,  y_test_pred)

    # Print nicely
    print("Training Performance:")
    print(f"- MAE : {model_train_mae:.4f}")
    print(f"- RMSE: {model_train_rmse:.4f}")
    print(f"- R2  : {model_train_r2:.4f}")

    print("\nTesting Performance:")
    print(f"- MAE : {model_test_mae:.4f}")
    print(f"- RMSE: {model_test_rmse:.4f}")
    print(f"- R2  : {model_test_r2:.4f}")

    # Store results for comparison
    results.append({
        "Model": name,
        "Train MAE": model_train_mae,
        "Train RMSE": model_train_rmse,
        "Train R2": model_train_r2,
        "Test MAE": model_test_mae,
        "Test RMSE": model_test_rmse,
        "Test R2": model_test_r2
    })

# Summary table
results_df = pd.DataFrame(results)
results_df
```
